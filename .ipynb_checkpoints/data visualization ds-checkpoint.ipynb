```python
#importing all the essential libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#extracting and reading the data
performance_data = pd.read_csv("book1.csv")
```

```python
performance_data
```

## Performance of Neha in 3 Weeks

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Performance',y='Neha',hue='Week',data = performance_data,palette = 'coolwarm')
plt.title("PERFORMANCE OF NEHA")
plt.show()
```

**We can use catplot as well**

```python
sns.set(style='darkgrid')
sns.catplot(x='Performance',y='Neha',data=performance_data,hue='Week')
plt.title("PERFORMANCE OF NEHA")
plt.show()
```

But we can see that barplot is way easier to visualize than scatter plot.

## Performance of Amit in 3 Weeks

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Performance',y='Amit',hue='Week',data = performance_data)
plt.title("PERFORMANCE OF AMIT")
plt.show()
```

By visualizing this plot, Amit is happy in all the three weeks with the Student Unicorn's contribution. Furthermore, he seems happy in first two weeks with the initiation of the Student Unicorn. Also, he is not much happy with identifying bottlenecks as compare to contribution and initiative.

**Using lineplot**

```python
sns.set(style='dark')
plt.figure(figsize=(7,4))
sns.lineplot(x='Performance',y='Amit',hue='Week',data = performance_data)
plt.title("PERFORMANCE OF AMIT")
plt.show()
```
